## Ladeinfrastruktur

```python
data_path = 'https://www.bundesnetzagentur.de/SharedDocs/Downloads/DE/Sachgebiete/Energie/Unternehmen_Institutionen/E_Mobilitaet/Ladesaeuleninfrastruktur.xlsx?__blob=publicationFile&v=20'

# Loading the excel sheet
df = pd.read_excel(data_path,
                   sheet_name='4.1 Ladepunkte je BL',
                   header=[6, 7],
                   index_col=4)

# Drop unnecessary columns
df = df.drop(columns=df.iloc[:, 0:4])
# Drop unnecessary index
df = df.drop(index='*NLP = Normalladepunkt, SLP = Schnellladepunkt')
# Rename the columns
df.columns.names = ['Date', 'Type']
```

```python
# Overview over cleaned table
df
```

```python
# Example for accessing a specific data element
df.loc['Bayern', (dt.datetime(2017, 7, 1) ,'gesamt')]
```

## Neuzulassungen


```python
data_path = 'https://www.kba.de/SharedDocs/Downloads/DE/Statistik/Fahrzeuge/FZ28/fz28_2022_01.xlsx?__blob=publicationFile'

# Loading the excel sheet
df = pd.read_excel(data_path,
                   sheet_name='FZ 28.9',
                   header=None,
                   index_col=1)

# Drop unnecessary rows
df = df.iloc[12:29]
# Just take colums for overall and electric-powered
df = df.iloc[:,[1, 4]]
# Rename the columns
df.columns = ['Insgesamt', 'Elektro-Antriebe']
# Rename the index
df.index.name = None
# Rename the first index value
idx = list(df.index)
idx[0] = 'Gesamt'
df.index = idx
```

```python
# Add a column for the percentage of electric powered cars
df['Anteil Elektro'] = df['Elektro-Antriebe'] / df['Insgesamt']
```

```python
# Overview of the cleaned table
df
```

```python
# Example for accessing one data element
df.loc['Bayern', 'Elektro-Antriebe']
```
